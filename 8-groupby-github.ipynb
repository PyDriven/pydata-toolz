# Groupby and Github

```python
from toolz.curried import *
import json
from glob import glob
file_pattern = 'data/github-january-2012/2012-01*.json'

# 有时github的数据是混乱的，包含奇怪的字符等...
# 很多时候 我们选择跳过它们，而不是处理它们
def try_load_json(s):
    try:
        return json.loads(s)
    except:
        return None
```

```python
records = pipe(file_pattern, glob, sorted, 
                             map(open), concat, 
                             map(try_load_json), filter(None))
```

```python
next(records)
```

```python
bytype = groupby(get('type'), take(100000, records))
```
